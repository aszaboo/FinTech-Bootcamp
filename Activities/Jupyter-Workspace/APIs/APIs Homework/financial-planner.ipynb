## Part 1 - Personal Finance Planner

### Collect Crypto Prices Using the `requests` Library

```python
# Set current amount of crypto assets
# YOUR CODE HERE!
my_btc = 1.2
my_eth = 5.3
```

```python
# Crypto API URLs
btc_url = "https://api.alternative.me/v2/ticker/Bitcoin/?convert=CAD"
eth_url = "https://api.alternative.me/v2/ticker/Ethereum/?convert=CAD"
```

```python
# Fetch current BTC price
# YOUR CODE HERE!
btc_data = requests.get(btc_url)
btc_content = btc_data.json()
btc_price = btc_content['data']['1']['quotes']['CAD']['price']
# Fetch current ETH price
# YOUR CODE HERE!
eth_data = requests.get(eth_url)
eth_content = eth_data.json()
eth_price = eth_content['data']['1027']['quotes']['CAD']['price']
# Compute current value of my crpto
# YOUR CODE HERE!
my_btc_value = my_btc * btc_price
my_eth_value = my_eth * eth_price
# Print current crypto wallet balance
print(f"The current value of your {my_btc} BTC is ${my_btc_value:0.2f}")
print(f"The current value of your {my_eth} ETH is ${my_eth_value:0.2f}")
```

### Collect Investments Data Using Alpaca: `SPY` (stocks) and `AGG` (bonds)

```python
# Set current amount of shares
my_agg = 200
my_spy = 50
```

```python
# Set Alpaca API key and secret
# YOUR CODE HERE!
alpaca_api_key = os.getenv('ALPACA_API_KEY')
alpaca_secret_key = os.getenv('ALPACA_SECRET_KEY')
# Create the Alpaca API object
# YOUR CODE HERE!
from alpaca_trade_api.rest import REST, TimeFrame
alpaca_api = REST(alpaca_api_key,alpaca_secret_key)
```

```python
# Format current date as ISO format
# YOUR CODE HERE!
import datetime
today = datetime.date.today()
today = pd.Timestamp(today, tz="America/New_York").isoformat()
# Set the tickers
tickers = ["AGG", "SPY"]

# Set timeframe to '1D' for Alpaca API
timeframe = "1D"

# Get current closing prices for SPY and AGG
# (use a limit=1000 parameter to call the most recent 1000 days of data)
# YOUR CODE HERE!
start_date = datetime.datetime.now() - datetime.timedelta(1000)
start_date = pd.Timestamp(start_date, tz="America/New_York").isoformat()

stocks_data_df = alpaca_api.get_bars(
    tickers,
    TimeFrame.Day,
    start = start_date,
    limit = 2000
).df
# Preview DataFrame
# YOUR CODE HERE!
stocks_data_df.tail()
```

```python
# Reorganize the DataFrame
# Separate ticker data
SPY = stocks_data_df[stocks_data_df['symbol']=='SPY'].drop('symbol', axis=1)
AGG = stocks_data_df[stocks_data_df['symbol']=='AGG'].drop('symbol', axis=1)
```

```python
# Concatenate the ticker DataFrames
stock_closes = pd.concat([SPY, AGG], axis=1, keys=["SPY", "AGG"])
```

```python
# turning the index into datetime
stock_closes.index = stock_closes.index.date
```

```python
# Pick AGG and SPY close prices
# YOUR CODE HERE!
agg_close_price = stock_closes['AGG']['close'][-1]
spy_close_price = stock_closes['SPY']['close'][-1]
# Print AGG and SPY close prices
print(f"Current AGG closing price: ${agg_close_price}")
print(f"Current SPY closing price: ${spy_close_price}")
```

```python
# Compute the current value of shares
# YOUR CODE HERE!
my_spy_value = spy_close_price * my_spy
my_agg_value = agg_close_price * my_agg
# Print current value of shares
print(f"The current value of your {my_spy} SPY shares is ${my_spy_value:0.2f}")
print(f"The current value of your {my_agg} AGG shares is ${my_agg_value:0.2f}")
```

### Savings Health Analysis

```python
# Set monthly household income
# YOUR CODE HERE!
monthly_income = 12000
# Consolidate financial assets data
# YOUR CODE HERE!
crypto_assets = my_btc_value + my_eth_value
trad_assets = my_spy_value + my_agg_value
# Create savings DataFrame
# YOUR CODE HERE!
savings = [crypto_assets,trad_assets]
df_savings = pd.DataFrame(savings)
df_savings.rename(columns={0: 'amount'}, inplace=True)
df_savings.index = ['crypto', 'shares']
# Display savings DataFrame
display(df_savings)
```

```python
# Plot savings pie chart
# YOUR CODE HERE!
df_savings.plot.pie(y="amount", title="Savings in Crypto vs. Shares")
```

```python
# Set ideal emergency fund
emergency_fund = monthly_income * 3

# Calculate total amount of savings
# YOUR CODE HERE!
# checking if amount is enough for emergency fund
total_savings = df_savings['amount'].sum()
# Validate saving health
# YOUR CODE HERE!
if total_savings > emergency_fund:
    print("Congratulations! You have enough money in your emergency fund.")
elif total_savings < emergency_fund:
    print(f"You need ${emergency_fund-total_savings} more to reach the ideal emergency fund amount")
```

## Part 2 - Retirement Planning

### Monte Carlo Simulation

```python
# Set start and end dates of five years back from today.
# Sample results may vary from the solution based on the time frame chosen
start_date2 = datetime.datetime.now() - datetime.timedelta(252 * 5)
start_date2 = pd.Timestamp(start_date2, tz="America/New_York").isoformat()
end_date2 = today
```

```python
# Get 5 years' worth of historical data for SPY and AGG
# (use a limit=1000 parameter to call the most recent 1000 days of data)
# YOUR CODE HERE!
history_data_df = alpaca_api.get_bars(
    tickers,
    TimeFrame.Day,
    start = start_date2,
    end = end_date2,
).df
```

```python
SPY_MC = stocks_data_df[stocks_data_df['symbol']=='SPY'].drop('symbol', axis=1)
AGG_MC = stocks_data_df[stocks_data_df['symbol']=='AGG'].drop('symbol', axis=1)
```

```python
# Concatenate the ticker DataFrames
stock_mc = pd.concat([SPY_MC, AGG_MC], axis=1, keys=["SPY", "AGG"])
# turning the index into datetime
stock_mc.index = stock_mc.index.date
```

```python
# Print the simulation input data
stock_mc.head()
```

```python
# Configuring a Monte Carlo simulation to forecast 30 years cumulative returns
# YOUR CODE HERE!
MC_thirtyyear = MCSimulation(
    portfolio_data = stock_mc,
    weights = [.5,.5],
    num_simulation = 500,
    num_trading_days = 252*30
)
```

```python
# Printing the simulation input data
# YOUR CODE HERE!
MC_thirtyyear.portfolio_data.head()
```

```python
# Running a Monte Carlo simulation to forecast 30 years cumulative returns
# YOUR CODE HERE!
MC_thirtyyear.calc_cumulative_return()
```

```python
# Plot simulation outcomes
# YOUR CODE HERE!
line_plot = MC_thirtyyear.plot_simulation()
```

```python
# Plot probability distribution and confidence intervals
# YOUR CODE HERE!
dist_plot = MC_thirtyyear.plot_distribution()
# Save the plot for future usage
dist_plot.get_figure().savefig('MC_thirtyyear_dist_plot.png',bbox_inches='tight')
```

### Retirement Analysis

```python
# Fetch summary statistics from the Monte Carlo simulation results
# YOUR CODE HERE!
sum_stat = MC_thirtyyear.summarize_cumulative_return()
# Print summary statistics
# YOUR CODE HERE!
print(sum_stat)
```

### Calculate the expected portfolio return at the `95%` lower and upper confidence intervals based on a `$20,000` initial investment.

```python
# Set initial investment
initial_investment = 20000

# Use the lower and upper `95%` confidence intervals to calculate the range of the possible outcomes of our $20,000
# YOUR CODE HERE!
ci_lower = round(sum_stat[8]*initial_investment,2)
ci_upper = round(sum_stat[9]*initial_investment,2)
# Print results
print(f"There is a 95% chance that an initial investment of ${initial_investment} in the portfolio"
      f" over the next 30 years will end within in the range of"
      f" ${ci_lower} and ${ci_upper}")
```

### Calculate the expected portfolio return at the `95%` lower and upper confidence intervals based on a `50%` increase in the initial investment.

```python
# Set initial investment
initial_investment2 = 20000 * 1.5

# Use the lower and upper `95%` confidence intervals to calculate the range of the possible outcomes of our $30,000
# YOUR CODE HERE!
ci_lower = round(sum_stat[8]*initial_investment2,2)
ci_upper = round(sum_stat[9]*initial_investment2,2)
# Print results
print(f"There is a 95% chance that an initial investment of ${initial_investment} in the portfolio"
      f" over the next 30 years will end within in the range of"
      f" ${ci_lower} and ${ci_upper}")
```
