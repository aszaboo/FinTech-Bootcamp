```python
# Load CSV data into Pandas using read_csv for each file.
invst_df = pd.read_csv("invstrs_leadership.csv")
america_df= pd.read_csv("fin_leaders_america.csv")
member_df= pd.read_csv("invstrs_leadership_members.csv")
fin_df = pd.read_csv("fin_leaders_members.csv")
```

```python
# Concatenate dues. Use the concat function to concatenate the fin_leaders_america.csv 
# data with the invstrs_leadership.csv data by axis='rows' and join='inner'.
dues_df = pd.concat(
    [invst_df, america_df], axis="rows", join="inner"
                   )
dues_df
```

```python
# Concatenate member data. Use the concat function to concatenate the fin_leaders_members.csv data
#and the invstrs_leadership_members.csv data by axis='columns' and join='inner'.
mem_df = pd.concat(
    [member_df,fin_df], axis="rows", join="inner",
        )
mem_df
```

```python
# You can mitigate this error 
mem_df.reset_index(inplace=True, drop=True)
```

```python
# Concatenate dues and member data. Use the concat function to concatenate 
# the two DataFrames made in steps 2 and 3 using by axis='columns' and join='inner'
dues_member_status = pd.concat(
    [dues_df, mem_df], axis="columns", join="inner"
)
dues_member_status
```

```python
# dropping duplicate rows
edit_df = dues_member_status.drop_duplicates()
edit_df
```
