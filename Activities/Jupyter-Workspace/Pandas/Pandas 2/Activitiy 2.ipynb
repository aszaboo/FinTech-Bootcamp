```python
import pandas as pd
df=pd.read_csv('bb_stock_data.csv')
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df
```

```python
# index data by Date.
df.set_index(['Date'], inplace=True)
```

```python
# Clean the data.
df.drop_duplicates()
```

```python
# Calculate daily returns.
df['Difference'] = df['Close'].diff()
df['pct_change'] = df['Close'].pct_change()
df.dropna()
```

```python
# Sort the DataFrame by Close in descending order using sort_values to obtain top daily return records.
df.sort_values(by= 'Close', ascending=False, inplace=True)
```

```python
df_top5 = df.iloc[:5]
```

```python
df_top5
```

```python
df_top5['pct_change'].plot(kind="bar")
```
