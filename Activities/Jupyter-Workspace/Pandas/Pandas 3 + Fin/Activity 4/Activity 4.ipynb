```python
shop = pd.read_csv("shop.csv",index_col="Date",parse_dates=True,infer_datetime_format=True)
```

```python
shop
```

```python
day_20 = shop['Close'].rolling(window=20).mean()
day_50 = shop['Close'].rolling(window=50).mean()
day_100 = shop['Close'].rolling(window=100).mean()
```

```python
import matplotlib.pyplot as plt

#plot each series
plt.plot(day_20, label='20', color='green')
plt.plot(day_50, label='50', color='steelblue')
plt.plot(day_100, label='100', color='purple')
plt.plot(shop['Close'], label='Close', color='red')
#add legend
plt.legend(title='Group')

#add axes labels and a title
plt.ylabel('Price', fontsize=14)
plt.xlabel('Time', fontsize=14)
plt.title('Closing Prices', fontsize=16)

#display plot
plt.show()
```

```python
std_20 = shop['Close'].rolling(window=20).std()
std_50 = shop['Close'].rolling(window=50).std()
std_100 = shop['Close'].rolling(window=100).std()
std = shop['Close'].std()
```

```python
plt.plot(std_20, label='20', color='green')
plt.plot(std_50, label='50', color='steelblue')
plt.plot(std_100, label='100', color='purple')
#add legend
plt.legend(title='Std')

#add axes labels and a title
plt.ylabel('Price', fontsize=14)
plt.xlabel('Time', fontsize=14)
plt.title('Closing Prices', fontsize=16)

#display plot
```
