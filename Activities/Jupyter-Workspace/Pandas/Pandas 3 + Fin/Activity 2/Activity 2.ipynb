```python
# Importing pandas and numpy
import pandas as pd
import numpy as np
!ls
```

```python
bmo = pd.read_csv('BMO.csv', index_col="date", parse_dates=True, infer_datetime_format=True)
cnq = pd.read_csv('CNQ.csv', index_col="date", parse_dates=True, infer_datetime_format=True)
cve = pd.read_csv('CVE.csv', index_col="date", parse_dates=True, infer_datetime_format=True)          
enb = pd.read_csv('ENB.csv', index_col="date", parse_dates=True, infer_datetime_format=True)
imo = pd.read_csv('IMO.csv', index_col="date", parse_dates=True, infer_datetime_format=True)
ipl = pd.read_csv('IPL.csv', index_col="date", parse_dates=True, infer_datetime_format=True)
trp = pd.read_csv('TRP.csv', index_col="date", parse_dates=True, infer_datetime_format=True)
```

```python
bmo
```

```python
# looking for na
bmo.dropna().copy()
```

```python
# Use the concat function to combine the seven DataFrames into a single combined DataFrame with an index of date
all_data = pd.concat([bmo,cnq,cve,enb,imo,ipl,trp], axis="columns", join="inner")
```

```python
all_data
```

```python
# Calculate the daily returns using the pct_change function.
returns = all_data.pct_change().dropna()
```

```python
returns
```

```python
# Use the corr function on the combined DataFrame to calculate and output a correlation table of each stock-to-stock pair.
correlation = returns.corr()
correlation
```
